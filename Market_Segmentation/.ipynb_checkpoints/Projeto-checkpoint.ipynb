## Criando modelo

```python
cat_features = ['region','reps','product']
```

```python
### Usanso CountEncoder para variaveis categóricas
count_enc = ce.CountEncoder()
count_encoded = count_enc.fit_transform(data[cat_features])
count_encoded.head()
```

```python
baseline = data.copy()
baseline = data.join(count_encoded.add_suffix('_count'))
baseline.head()
```

```python
### Adicionando o preco unitario
baseline['unit_price'] = baseline['revenue']/baseline['qty']
baseline.head()
```

```python
baseline.drop(cat_features,axis=1,inplace=True)
```

```python
## Escolhendo as features que serão usadas
features = ['qty','revenue','region_count','unit_price']
```

```python
baseline = pd.DataFrame(baseline[features])
baseline.head()
```

####  Será usado o método kMeans, para escolha do número de cluster foi usado o método do cotovelo e a análise da silhueta

```python
## Método do cotovelo
baseline_array = baseline.values
inercia = []

for i in range(2,9):
    modelo = KMeans(n_clusters=i, random_state=1)
    modelo.fit_predict(baseline_array)
    inercia.append((i,modelo.inertia_,))
```

```python
### Aqui podemos perceber que o numero de cluster bom poderia ser 4 pois o erro cai abruptamente e começa a se "estabilizar"
plt.plot([w[0] for w in inercia],[w[1] for w in inercia], marker="X")
```

```python
## Pela análise da silhueta o melhor número de cluster é 4
num_clusters = [2, 3, 4, 5, 6, 7, 8] 
  
for n_clusters in num_clusters: 
  
    cluster = KMeans(n_clusters = n_clusters) 
    cluster_labels = cluster.fit_predict(baseline_array)  
    silhouette_avg = silhouette_score(baseline_array, cluster_labels)
        
    print("For no of clusters =", n_clusters, 
          " The average silhouette_score is :", silhouette_avg) 
```

```python
modelo = KMeans(n_clusters=4)
baseline['cluster'] = modelo.fit_predict(baseline_array)
```

```python
data['cluster'] = baseline['cluster']
data.to_csv('market_segmentation.csv')
```

```python
def nome_regiao(x):
    for i,linha in baseline.iterrows():
        if int(linha['region_count']) == x:
            return data.loc[i,'region']
```

```python
def nome_rep(x):
    for i,linha in baseline.iterrows():
        if int(linha['reps_count']) == x:
            return data.loc[i,'reps']
```
